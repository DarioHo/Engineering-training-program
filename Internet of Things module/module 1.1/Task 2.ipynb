```python
# import libraries
import pandas as pd
```

```python
# read a CSV format text file
# the data is sorted by utc_time
df = pd.read_csv('beijing_air_quality_data.csv', parse_dates=['utc_time'])
```

```python
# for temperature and humidity
# calculate its mean, standard deviation

# temperature mean, standard deviation
mean_temperature = df['temperature'].mean()
std_temperature = df['temperature'].std()

print ('The mean of temperature is', mean_temperature)
print ('The standard divation of temperature is', std_temperature)

# humidity mean, standard deviation
mean_humidity = df['humidity'].mean()
std_humidity = df['humidity'].mean()

print ('\nThe mean of humidity is', mean_humidity)
print ('The standard divation of humidity', std_humidity)
```

```python
# how many unique site_id?
# Hint: check pd.unique method

site_id = df['site_id'].unique()

# find the total unique numbers
site_number = df['site_id'].nunique()
print ('The number of unique numbers are:', site_number)

# sort the unique site_id for easier reading
site_id.sort()
print (site_id)
```

```python
# what is the range of utc_time?
# Hint: you can call min() and max() for the utc_time column

max_utc_time = df['utc_time'].max()
min_utc_time = df['utc_time'].min()

print ('The range of utc_time is', min_utc_time, 'to', max_utc_time)
```

```python
# for pm25, fill NAs with mean value

mean_pm25 = df['pm25'].mean()

print (mean_pm25)
df.fillna(mean_pm25)
```

```python
# select data for a station where site_id equals to dongsi_aq
dongsi_df = df[df['site_id'] == 'dongsi_aq']

dongsi_df_pm25 = dongsi_df['pm25']

# calculate the mean value of pm25 at this station

mean_humidity = dongsi_df_pm25.mean()
print ('The mean value of pm25 at the station dongsi_aq is', mean_humidity)

# calculate how many data points when pm25 >= 50 at this station
# Hint: check count() method

print ('\nData points when pm25 >= 50 is:')
dongsi_df_pm25[dongsi_df_pm25 >=50].count()
```

```python
# how many NAs for pm25 values at this station
# Hint: check isnull() 
print ('NAs for pm25 values at this station is:')
dongsi_df_pm25.isnull().sum()
```
