```python
# import libraries
import numpy as np
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a normal (Gaussian) distribution
# between -20 and 20 with .001 step size
x = np.arange(-20, 20, 0.001)
# mean
u = 2
# standard deviation
sigma = 10
# probability density function for normal distribution
y = scipy.stats.norm.pdf(x, u, sigma)
```

```python
# plot the normal distribution
plt.plot(x,y)
# to save the image in the folder
plt.savefig('normal_distribution,png')
```

```python
# load CSV data into Pandas DataFrame
df = pd.read_csv('beijing_air_quality_data.csv', parse_dates=['utc_time'])
```

```python
# Prepare the data for plotting
dongsi_aq = df[df['site_id'] == 'dongsi_aq']
dongsi_aq = dongsi_aq[['site_id', 'utc_time', 'temperature']]
dongsi_aq
```

```python
# plot temperature series
dongsi_aq.plot(x='utc_time', y='temperature')
plt.show()
plt.savefig('plot_series.png')
```

```python
# plot temperature histogram
# Hint: call hist()

temper_list = df['temperature'].tolist()
plt.hist(temper_list)
plt.show()
plt.savefig('hist.png')
```
